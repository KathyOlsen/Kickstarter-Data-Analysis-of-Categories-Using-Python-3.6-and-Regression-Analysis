```python
import pandas as pd

Location = "Kickstartercombo.csv"
df = pd.read_csv(Location)
df.head()
```

```python
df.dtypes
```

```python
df.columns = ['Unnamed: 0','ID','name','main_category','category','country','state','goal','pledged','currency','launched','deadline','backers','temp usd pledged','usd_goal_real','usd_pledged_real']
df.head()
```

```python
df.count()
```

```python
pd.value_counts(df['state'])
```

```python
dfdrop = df
dfdrop = dfdrop.drop('Unnamed: 0', axis=1)
dfdrop = dfdrop.drop('usd_goal_real', axis=1)
dfdrop = dfdrop.drop('usd_pledged_real', axis=1)
dfdrop = dfdrop.dropna(axis=0,subset=['temp usd pledged'])
dfdrop = dfdrop.drop(dfdrop[dfdrop['state']=='live'].index)
dfdrop = dfdrop.drop(dfdrop[dfdrop['state']=='undefined'].index)
dfdrop = dfdrop.sort_values(by='ID',ascending=1)
dfdrop.head()
```

```python
print(len(dfdrop.drop_duplicates(['ID'])))
```

```python
dfdrop = dfdrop.drop_duplicates(['ID'], keep='first')
dfdrop.head()
```

```python
dfdrop.count()
```

```python
dfdrop = dfdrop.drop_duplicates(['name'], keep='first')
dfdrop.head()
```

```python
dfdrop.count()
```

```python
import numpy as np
dfdrop['state_recat'] = np.where(dfdrop['pledged']>=dfdrop['goal'],'successful','not_successful')
dfdrop.head()
```

```python
## Drop rows where state_recat = 'successful' and backers = 0 (state should have been classified as 'undefined').
dfdrop = dfdrop.drop(dfdrop[(dfdrop['state_recat']=='successful') & (dfdrop['backers']==0)].index)
dfdrop.head()
```

```python
## Create numeric variable to represent successful or not_successful.
def score_to_numeric(x):
    if x=='sucessful':
        return 1
    if x=='not_successful':
        return 0
```

```python
dfdrop['success_val'] = dfdrop['state_recat'].apply(score_to_numeric)
dfdrop_success_val = pd.get_dummies(dfdrop['state_recat'])
dfdrop_success_val.head()
```

```python
dfdrop_new = dfdrop.join(dfdrop_success_val)
dfdrop_new.head()
```

```python
print(len(dfdrop_new[(dfdrop_new['currency']=='USD') & ((dfdrop_new['pledged'])!=(dfdrop_new['temp usd pledged']))]))
```

```python
dfdrop_new['usd pledged'] = np.where((dfdrop_new['currency']=='USD') & (dfdrop_new['pledged']!=dfdrop_new['temp usd pledged']),dfdrop_new['pledged'],dfdrop_new['temp usd pledged'])
dfdrop_new.head()
```

```python
dfdrop_new = dfdrop_new.drop('temp usd pledged', axis=1)
```

```python
pd.value_counts(dfdrop_new['currency'])
```

```python
print(len(dfdrop_new[dfdrop_new['usd pledged']==0]))
```

```python
print(len(dfdrop_new[(dfdrop_new['currency']=='USD') & (dfdrop_new['usd pledged']==0)]))
```

```python
print(len(dfdrop_new[(dfdrop_new['currency']!='USD') & (dfdrop_new['usd pledged']==0)]))
```

```python
## Create column to represent USD or not_USD
dfdrop_new['USD_recat'] = np.where(dfdrop_new['currency']=='USD','USD','not_USD')
dfdrop_new.head()
```

```python
## Create numeric variable to represent USD or not_USD.
def score_to_numeric2(x):
    if x=='USD':
        return 1
    if x=='not_USD':
        return 0
```

```python
dfdrop_new['USD_val'] = dfdrop_new['USD_recat'].apply(score_to_numeric2)
dfdrop_USD_val = pd.get_dummies(dfdrop_new['USD_recat'])
dfdrop_USD_val.head()
```

```python
dfdrop_new2 = dfdrop_new.join(dfdrop_USD_val)
dfdrop_new2.head()
```

```python
## Create new column for ratio of pledged to goal.

dfdrop_new2['ratio'] = (dfdrop_new2['pledged'])/(dfdrop_new2['goal'])
dfdrop_new2.head()
```

```python
## Adjust ratio to facilitate creation of new 'usd goal' column.
def x_adjusted(x):
    if x==0:
        return 0.00001
    else:
        return x
```

```python
dfdrop_new2['adjusted ratio'] = dfdrop_new2['ratio'].apply(x_adjusted)
dfdrop_new2.head()
```

```python
dfdrop_new2['usd goal temp'] = (dfdrop_new2['usd pledged'])/(dfdrop_new2['adjusted ratio'])
dfdrop_new2.head()
```

```python
## Create column with goal only if in usd and goals in all other currencies = 0.
dfdrop_new2['goal if usd'] = dfdrop_new2['goal']*dfdrop_new2['USD']
dfdrop_new2.head()
```

```python
## Create new column to substitute 'goal if  usd' for those rows in 'usd goal temp' column that = 0 because pledged = 0.
dfdrop_new2['usd goal temp_2'] = np.where(dfdrop_new2['goal if usd']>dfdrop_new2['usd goal temp'],dfdrop_new2['goal if usd'],dfdrop_new2['usd goal temp'])
dfdrop_new2.head()
```

```python
print(len(dfdrop_new2[(dfdrop_new2['usd goal temp_2']==0) & (dfdrop_new2['usd pledged']==0)]))
```

```python
## Create column to represent GBP or not_GBP
dfdrop_new2['GBP_recat'] = np.where(dfdrop_new2['currency']=='GBP','GBP','not_GBP')
dfdrop_new2.head()
```

```python
## Create numeric variable to represent GBP or not_GBP.
def score_to_numeric3(x):
    if x=='GBP':
        return 1
    if x=='not_GBP':
        return 0
```

```python
dfdrop_new2['GBP_val'] = dfdrop_new2['GBP_recat'].apply(score_to_numeric3)
dfdrop_GBP_val = pd.get_dummies(dfdrop_new2['GBP_recat'])
dfdrop_GBP_val.head()
```

```python
dfdrop_new3 = dfdrop_new2.join(dfdrop_GBP_val)
dfdrop_new3.head()
```

```python
## Create column with goal in usd (using today's currency conversion) only if originally in gbp 
## and goals in all other currencies = 0.
dfdrop_new3['usd goal if gbp'] = dfdrop_new3['goal']*dfdrop_new3['GBP']*1.32013
dfdrop_new3.head()
```

```python
## Create new column to substitute 'usd goal if  gbp' for those from 'usd goal temp' column that = 0 because pledged = 0.
dfdrop_new3['usd goal temp_3'] = np.where((((dfdrop_new3['usd goal if gbp'])>0) & (dfdrop_new3['usd goal temp_2']==0)),dfdrop_new3['usd goal if gbp'],dfdrop_new3['usd goal temp_2'])
dfdrop_new3.head()
```

```python
print(len(dfdrop_new3[(dfdrop_new3['usd goal temp_3']==0) & (dfdrop_new3['usd pledged']==0)]))
```

```python
## Create column to represent EUR or not_EUR
dfdrop_new3['EUR_recat'] = np.where(dfdrop_new3['currency']=='EUR','EUR','not_EUR')
dfdrop_new3.head()
```

```python
## Create numeric variable to represent EUR or not_EUR.
def score_to_numeric4(x):
    if x=='EUR':
        return 1
    if x=='not_EUR':
        return 0
```

```python
dfdrop_new3['EUR_val'] = dfdrop_new3['EUR_recat'].apply(score_to_numeric4)
dfdrop_EUR_val = pd.get_dummies(dfdrop_new3['EUR_recat'])
dfdrop_EUR_val.head()
```

```python
dfdrop_new4 = dfdrop_new3.join(dfdrop_EUR_val)
dfdrop_new4.head()
```

```python
## Create column with goal in usd (using today's currency conversion) only if originally in eur 
## and goals in all other currencies = 0.
dfdrop_new4['usd goal if eur'] = dfdrop_new4['goal']*dfdrop_new4['EUR']*1.16687
dfdrop_new4.head()
```

```python
## Create new column to substitute 'usd goal if  eur' for those from 'usd goal temp' column that = 0 because pledged = 0.
dfdrop_new4['usd goal temp_4'] = np.where((((dfdrop_new4['usd goal if eur'])>0) & (dfdrop_new4['usd goal temp_3']==0)),dfdrop_new4['usd goal if eur'],dfdrop_new4['usd goal temp_3'])
dfdrop_new4.head()
```

```python
print(len(dfdrop_new4[(dfdrop_new4['usd goal temp_4']==0) & (dfdrop_new4['usd pledged']==0)]))
```

```python
## Create column to represent CAD or not_CAD
dfdrop_new4['CAD_recat'] = np.where(dfdrop_new4['currency']=='CAD','CAD','not_CAD')
dfdrop_new4.head()
```

```python
## Create numeric variable to represent CAD or not_CAD.
def score_to_numeric5(x):
    if x=='CAD':
        return 1
    if x=='not_CAD':
        return 0
```

```python
dfdrop_new4['CAD_val'] = dfdrop_new4['CAD_recat'].apply(score_to_numeric5)
dfdrop_CAD_val = pd.get_dummies(dfdrop_new4['CAD_recat'])
dfdrop_CAD_val.head()
```

```python
dfdrop_new5 = dfdrop_new4.join(dfdrop_CAD_val)
dfdrop_new5.head()
```

```python
## Create column with goal in usd (using today's currency conversion) only if originally in cad 
## and goals in all other currencies = 0.
dfdrop_new5['usd goal if cad'] = dfdrop_new5['goal']*dfdrop_new5['CAD']*0.75978
dfdrop_new5.head()
```

```python
## Create new column to substitute 'usd goal if cad' for those from 'usd goal temp' column that = 0 because pledged = 0.
dfdrop_new5['usd goal temp_5'] = np.where(((dfdrop_new5['usd goal if cad']>0) & (dfdrop_new5['usd goal temp_4']==0)),dfdrop_new5['usd goal if cad'],dfdrop_new5['usd goal temp_4'])
dfdrop_new5.head()
```

```python
print(len(dfdrop_new5[(dfdrop_new5['usd goal temp_5']==0) & (dfdrop_new5['usd pledged']==0)]))
```

```python
## Create column to represent AUD or not_AUD
dfdrop_new5['AUD_recat'] = np.where(dfdrop_new5['currency']=='AUD','AUD','not_AUD')
dfdrop_new5.head()
```

```python
## Create numeric variable to represent AUD or not_AUD.
def score_to_numeric6(x):
    if x=='AUD':
        return 1
    if x=='not_AUD':
        return 0
```

```python
dfdrop_new5['AUD_val'] = dfdrop_new5['AUD_recat'].apply(score_to_numeric6)
dfdrop_AUD_val = pd.get_dummies(dfdrop_new5['AUD_recat'])
dfdrop_AUD_val.head()
```

```python
dfdrop_new6 = dfdrop_new5.join(dfdrop_AUD_val)
dfdrop_new6.head()
```

```python
## Create column with goal in usd (using today's currency conversion) only if originally in aud 
## and goals in all other currencies = 0.
dfdrop_new6['usd goal if aud'] = dfdrop_new6['goal']*dfdrop_new6['AUD']*0.74050
dfdrop_new6.head()
```

```python
## Create new column to substitute 'usd goal if aud' for those from 'usd goal temp' column that = 0 because pledged = 0.
dfdrop_new6['usd goal'] = np.where(((dfdrop_new6['usd goal if aud']>0) & (dfdrop_new6['usd goal temp_5']==0)),dfdrop_new6['usd goal if aud'],dfdrop_new6['usd goal temp_5'])
dfdrop_new6.head()
```

```python
print(len(dfdrop_new6[(dfdrop_new6['usd goal']==0) & (dfdrop_new6['usd pledged']==0)]))
```

```python
dfdrop_new6 = dfdrop_new6.drop(dfdrop_new6[(dfdrop_new6['currency']!='USD') & (dfdrop_new6['currency']!='GBP') & (dfdrop_new6['currency']!='EUR') & (dfdrop_new6['currency']!='CAD') & (dfdrop_new6['currency']!='AUD')  & (dfdrop_new6['usd pledged']==0)].index)
dfdrop_new6.head()
```

```python
dfdrop_new6.count()
```

```python
df2 = dfdrop_new6
df2 = df2.drop('usd goal temp', axis=1)
df2 = df2.drop('usd goal temp_2', axis=1)
df2 = df2.drop('usd goal temp_3', axis=1)
df2 = df2.drop('usd goal temp_4', axis=1)
df2 = df2.drop('usd goal temp_5', axis=1)
df2 = df2.drop('goal if usd', axis=1)
df2 = df2.drop('usd goal if gbp', axis=1)
df2 = df2.drop('usd goal if eur', axis=1)
df2 = df2.drop('usd goal if cad', axis=1)
df2 = df2.drop('usd goal if aud', axis=1)
df2 = df2.drop('success_val', axis=1)
df2 = df2.drop('USD_recat', axis=1)
df2 = df2.drop('GBP_recat', axis=1)
df2 = df2.drop('EUR_recat', axis=1)
df2 = df2.drop('CAD_recat', axis=1)
df2 = df2.drop('AUD_recat', axis=1)
df2 = df2.drop('USD_val', axis=1)
df2 = df2.drop('GBP_val', axis=1)
df2 = df2.drop('EUR_val', axis=1)
df2 = df2.drop('CAD_val', axis=1)
df2 = df2.drop('AUD_val', axis=1)
df2 = df2.drop('GBP', axis=1)
df2 = df2.drop('not_GBP', axis=1)
df2 = df2.drop('EUR', axis=1)
df2 = df2.drop('not_EUR', axis=1)
df2 = df2.drop('CAD', axis=1)
df2 = df2.drop('not_CAD', axis=1)
df2 = df2.drop('AUD', axis=1)
df2 = df2.drop('not_AUD', axis=1)
df2 = df2.drop('adjusted ratio', axis=1)
df2.head()
```

```python
df2.columns = ['ID','name','main_category','category','country','state','goal','pledged','currency','launched','deadline','backers','state_recat','not_successful','successful','usd_pledged','USD','not_USD','ratio','usd_goal']
df2.head()
```
